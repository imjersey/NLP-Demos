```python
#encoding=utf-8
#Author = Lijianquan
#date= 2017/9/27
#funtion : Word2Vector
from gensim.models import word2vec
#import log_config
import os
import codecs
import sys
#logger = log_config.getLogging("w2v")
def ReadStopWords(file_name):
    stop_d = {}
    with open(file_name, encoding='utf-8') as f:
        yw_stopwords = f.read().split('\n')
    for seg in yw_stopwords:
        stop_d[seg] = 0
    return stop_d
def ReadModel(model_mj):
    model_mj.init_sims(replace=True)
    segment_data_file = './data/cv.txt'
    with open(segment_data_file, 'w') as fp:
        i = 0
        for k in model_mj.wv.index2word:
            if model_mj.__contains__(k):
               try:
                    fp.write(k + ',')
                    t_l = list(model_mj[k])
                    for i in range(299):
                        fp.write(str(t_l[i]) + ',')
                    fp.write(str(t_l[299]))
                    fp.write('\n')
                    i += 1
                    if i >= 10000:
                        print("Write 10000")
                        i = 0
               except Exception as e:
                   print(e)
def CutSymbol(query,yw_stopwords):
    line = []
    for seg in query:
        if seg in yw_stopwords:
            pass
        else:
            line.append(seg)
    return line
def open_txt(file_name):
    with open(file_name,'r',encoding='utf-8') as f:
        while True:
            line = f.readline()
            if not line:
                return
            yield line.strip()
def ReadTxt(file_name,yw_stopwords):
    r_lines = []
    for line in open_txt(file_name):
        r_line = CutSymbol(line,yw_stopwords)
        r_lines.append(r_line)
    return r_lines
def PrepareData1(yw_stopwords):
    copus_path = './copus-back/'
    r_lines = []
    # copus_path = './data/'
    files = os.listdir(copus_path)
    # print (files)
    # sys.exit()
    for fi in files:
        fi_d = os.path.join(copus_path, fi)
        # print(fi_d)
        try:
            r = ReadTxt(fi_d,yw_stopwords)
            r_lines.extend(r)
        except Exception as e:
            print(e)
        print('ready'+ fi_d)
        logger.info('ready'+ fi_d)
    for i in range(100):
        try:
            r = ReadTxt('./data/mojiezhitou.txt',yw_stopwords)
            r_lines.extend(r)
        except Exception as e:
            print(e)
    return r_lines
def PrepareData(yw_stopwords):
    file_path = './copus-back'
    segment_data_file = './data/mj.txt'
    with open(segment_data_file, 'w') as fp:
        for dirpath, _, filenames in os.walk(file_path):
            for filename in (x for x in sorted(filenames) if x.endswith('.txt')):
                db_path = os.path.join(dirpath, filename)
                print(db_path)
                logger.info(db_path)
                with codecs.open(db_path, 'r', 'utf-8', errors='ignore') as f:
                    print("Start read")
                    for line in f:
                        correct_line = line.split('。')
                        for line in correct_line:
                            r_line = CutSymbol(line, yw_stopwords)
                            # words = thu.cut(line, text=False)
                            for character in r_line:
                                    fp.write(character + ' ')
                        fp.write('\n')
                    print("The end segmentation of {}".format(db_path))
                    logger.info("The end segmentation of {}".format(db_path))
            mj_path = './data/mojiezhitou.txt'
            for i in range(100):
                with codecs.open(mj_path, 'r', 'utf-8', errors='ignore') as f:
                    print("Start read")
                    for line in f:
                        correct_line = line.split('。')
                        for line in correct_line:
                            r_line = CutSymbol(line, yw_stopwords)
                            # words = thu.cut(line, text=False)
                            for character in r_line:
                                    fp.write(character + ' ')
                        fp.write('\n')
                    print("The end segmentation of {}".format(mj_path))
                    logger.info("The end segmentation of {}".format(mj_path))

def Character2Vector():
    yw_stopwords = ReadStopWords("./data/yw_StopWords.txt")
    PrepareData(yw_stopwords)
    print('ready')
    logger.info('ready')
    sentences = word2vec.Text8Corpus('./data/mj.txt')
    model = word2vec.Word2Vec(sentences, size=300, min_count=10)
    print('success')
    logger.info('success')
    model.save('./CharacterVector/character.model')
if __name__ == '__main__':
    # Character2Vector()
    # yw_stopwords = ReadStopWords("./data/yw_StopWords.txt")
    # r = ReadTxt('./data/mojiezhitou.txt',yw_stopwords)
    # model = word2vec.Word2Vec(r, size=100, window=5, min_count=1, workers=4)
    # wv_ls = ReadModel(model)
    # print(wv_ls[0])
    # model.save('./CharacterVector/character.bin')
    model = word2vec.Word2Vec.load('./character.model')
    #ReadModel(model)
    
    





```
