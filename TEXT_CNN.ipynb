```python
#get dicts
dict = {}
q2a = {}
MAX_LEN = 0
VEC_LEN = 200
Curcate = ""
for rowx in range(0, sheet.nrows):
    cols = sheet.row_values(rowx)
    MAX_LEN = max(MAX_LEN, len(cols[2]), len(cols[1]) )
    
    if cols[2] != '':
        Curcate = cols[2]
        dict[cols[2]] = []
        dict[cols[2]].append(cols[1])
    else :
        dict[Curcate].append(cols[1])
        
    q2a[cols[1] ] = Curcate
```

```python
output = []

for key in dict:
    for i in range(len(dict[key])):
        for j in range(i+1,len(dict[key])):
            tmp = (dict[key][i], dict[key][j], 1)
            output.append(tmp)
    
```
